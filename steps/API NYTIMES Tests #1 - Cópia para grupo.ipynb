```python
#normalizing what is the list of dictionary
df_overview_normalized = pd.json_normalize(response3.json())
```

```python
#showing what is inside the results.lists
df_overview_normalized['results.lists'][0]
```

```python
#normalizing results.list for all the lists available in Best Seller Lists
allListsBestSeller = df_overview_normalized['results.lists'][0]
```

```python
#create a dataframe with the normalized data and put it into a variable
showAllListsBestSeller = pd.DataFrame(allListsBestSeller)
showAllListsBestSeller
```
